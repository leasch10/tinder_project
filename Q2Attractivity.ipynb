```python
import pandas as pd 
df = pd.read_csv("Speed+Dating+Data.csv", encoding='latin1')
df.head()
```

```python
import plotly.io as pio
import plotly.express as px
import plotly.graph_objects as go
```

```python
#Ã©valuation de l'attractivitÃ© lors du sign-up 
print(df["attr1_1"].mean())

fig = go.Figure()

# Add histogram for match == 0
fig.add_trace(go.Histogram(x=df["attr1_1"], name='Importance of Attractivity', opacity=0.5))

# Add histogram for match == 1
fig.add_trace(go.Histogram(x=df["attr1_2"], name='Same sexe evaluation', opacity=0.5))

# Update layout
fig.update_layout(
    barmode='overlay',
    title='Superimposed Histograms of sign-up question ',
    xaxis_title='attr1_s',
    yaxis_title='Frequency',
    legend_title='Match Condition',
    template='plotly_white'
)

# Show the figure
fig.show()
```

```python
fig = px.box(df, x= "match", y = "attr")
fig.show()
print(df["attr"].mean())
```

```python
#on note qu'Ã  mi chemin, l'attractivitÃ© n'est plus si importante vs lors de l'Ã©valuation du match) 
fig = px.box(df, x= "match", y = "attr1_s")
fig.show()
```

```python
fig = go.Figure()

# Add histogram for match == 0
fig.add_trace(go.Histogram(x=df[df["match"] == 0]["attr1_s"], name='Match = 0', opacity=0.5))

# Add histogram for match == 1
fig.add_trace(go.Histogram(x=df[df["match"] == 1]["attr1_s"], name='Match = 1', opacity=0.5))

# Update layout
fig.update_layout(
    barmode='overlay',
    title='Superimposed Histograms of attr1_s for Match = 0 and Match = 1',
    xaxis_title='attr1_s',
    yaxis_title='Frequency',
    legend_title='Match Condition',
    template='plotly_white'
)

# Show the figure
fig.show()
print(df["attr1_s"].mean())
```
