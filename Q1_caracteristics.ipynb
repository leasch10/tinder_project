```python
#faire la moyenne de la question 1 pour visualiser plus facilement l'importance des attributs 

attr1 = df['mean_attr_importance'] = df[['attr1_1', 'attr1_2', 'attr1_3']].mean(axis=1)
sinc1= df['mean_sinc_importance'] = df[['sinc1_1', 'sinc1_2', 'sinc1_3']].mean(axis=1)
intel1= df['mean_intel_importance'] = df[['intel1_1', 'intel1_2', 'intel1_3']].mean(axis=1)
fun1 = df['mean_fun_importance'] = df[['fun1_1', 'fun1_2', 'fun1_3']].mean(axis=1)
amb1 = df['mean_amb_importance'] = df[['amb1_1', 'amb1_2', 'amb1_3']].mean(axis=1)
shar1 = df['mean_shar_importance'] = df[['shar1_1', 'shar1_2', 'shar1_3']].mean(axis=1)
```

```python
import plotly.express as px
import plotly.graph_objects as go
fig = go.Figure()

fig.add_trace(go.Histogram(x=df['mean_attr_importance'], name='Attr Importance', opacity=0.5))
fig.add_trace(go.Histogram(x=df['mean_sinc_importance'], name='Sinc Importance', opacity=0.5))
fig.add_trace(go.Histogram(x=df['mean_intel_importance'], name='Intel Importance', opacity=0.5))
fig.add_trace(go.Histogram(x=df['mean_fun_importance'], name='Fun Importance', opacity=0.5))
fig.add_trace(go.Histogram(x=df['mean_amb_importance'], name='Amb Importance', opacity=0.5))
fig.add_trace(go.Histogram(x=df['mean_shar_importance'], name='Shar Importance', opacity=0.5))

# Update layout
fig.update_layout(
    barmode='overlay',
    title='Superimposed Histogram of Various Importance Scores',
    xaxis_title='Importance Scores',
    yaxis_title='Count',
    legend_title='Importance Type(Mean)',
    template='plotly_white'
)

# Show the figure
fig.show()
```
